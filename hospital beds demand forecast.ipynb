## Nur-Sultan dataset

```python
#Assemble Nur-sultan dataset
data=pd.DataFrame(ds_demography['Нур-Султан'])
data.rename(columns={"Нур-Султан": "Население"}, inplace=True)
data['Заболеваемость']=ds_illness['Нур-Султан']
data['Госпитализации']=ds_hospitalized['Нур-Султан']
data['Койки']=ds_beds['Нур-Султан']
data
```

```python
#dropna
data.dropna(inplace=True)
data
```

# Calculate the demand

```python
#TODo insert link to the methodology
```

```python
#getting ill dendency
gi_tend=data['Заболеваемость'][1:].values/data['Заболеваемость'][:-1].values
gi_tend.shape
```

```python
#getting ill dendency
hosp_tend=data['Госпитализации'][1:].values/data['Госпитализации'][:-1].values
hosp_tend#.shape
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
#calculate moving averages
ma_gi_tend=moving_average(gi_tend, 3)
ma_hosp_tend=moving_average(hosp_tend, 3)
```

```python
shift=data.shape[0]-ma_gi_tend.shape[0]
```

```python
#align calculated tendencies with the data
a=np.zeros(data.shape[0])
a[shift+1:]=ma_hosp_tend[:-1]
ma_hosp_tend=a
ma_hosp_tend
```

```python
b=np.zeros(data.shape[0])
b[shift+1:]=ma_gi_tend[:-1]
ma_gi_tend=b
ma_gi_tend
```

```python
#Demand
BED_TURNOVER=30# how many times 1hospital bed is used per year, given that on average it is occupied 10 days per patient

pop_f=data['Население'].values[1:]
coef_gi=data['Заболеваемость'].values[:-1]/data['Население'].values[:-1]*ma_gi_tend[:-1]
coef_hosp=data['Госпитализации'].values[:-1]/data['Заболеваемость'].values[:-1]*ma_hosp_tend[:-1]
demand_beds=pop_f*coef_gi*coef_hosp/BED_TURNOVER
demand_beds
```

```python
data['Спрос на койки']=0
data['Спрос на койки'].values[1:]=demand_beds
```

## Build regression model population vs. demand

```python
from sklearn.model_selection import train_test_split

#models to test
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score
```

```python
X=data.iloc[5:,:]['Население'].values.reshape(-1, 1)
Y=data.iloc[5:,:]['Спрос на койки'].astype('float64').values
```

```python
# estimate 95% confidence interval on error

# NOTE: based on conversation on stackexchange: 
# https://stats.stackexchange.com/questions/247551/how-to-determine-the-confidence-of-a-neural-network-prediction
# towards bottom of the page.

from math import sqrt

def error_conf(error, n):
    term = 1.96*sqrt((error*(1-error))/n)
    lb = error - term
    ub = error + term
    
    return lb, ub

# estimate 95% confidence interval

n = 1000
test_acc=0.97
lb, ub = error_conf(1-test_acc, n)

print("95% confidence interval: {:.2f}%-{:.2f}%".format((1-ub)*100,(1-lb)*100))
```
